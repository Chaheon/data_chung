## 데이터 시각화 

### matplotlib: 벡터친화적인 시각화 

#### 예제1 

```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4]
y=[1,2,3,2]
```

```python
plt.plot(x,y,'x')
```

```python
y2=[1.1,2.1,3.2,1]
```

```python
plt.plot(x,y,':o')
plt.plot(x,y2,':o')
```

#### 예제2

```python
import pandas as pd 
```

```python
dfdata=pd.read_csv("dfdata.csv")
```

```python
plt.plot(dfdata.toeic,dfdata.gpa,'o')
```

```python
toeic_y=dfdata.query('employed=="Y"').toeic
toeic_n=dfdata.query('employed=="N"').toeic
gpa_y=dfdata.query('employed=="Y"').gpa
gpa_n=dfdata.query('employed=="N"').gpa
plt.plot(toeic_y,gpa_y,'o')
plt.plot(toeic_n,gpa_n,'o')
```

데이터프레임 -> 쿼리문 -> 필터링된 데이터 프레임 -> 벡터화 -> 저장 -> 플랏 

### seaborn: 데이터프레임 친화적인 시각화 

#### 예제2 (이어서)

```python
import seaborn as sns 
```

```python
sns.relplot(data=dfdata, x='toeic',y='gpa',hue='employed')
```

`–` 취업된 사람들이 각각 어떠한 회사에 갔는지 궁금하다. 

```python
sns.relplot(data=dfdata.query('employed=="Y"'), x='toeic',y='gpa',hue='company')
```

`-` 연봉정보도 한눈에 알아보기 쉽게 그리고 싶다면?  

```python
sns.relplot(data=dfdata.query('employed=="Y"'), x='toeic',y='gpa',hue='company',size='salary')
```
