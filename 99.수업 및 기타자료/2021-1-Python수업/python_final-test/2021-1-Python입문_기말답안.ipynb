```python
# 패키지 

import numpy as np
import pandas as pd
```

```python
class JPsord:
    
    def __init__(self, prob , user = 100, d = 0): # 사용할 변수를 선언하고 초기화하는 용도. 
        self.prob = prob 
        self.user = user
        self.d = d
        
        self.attemptresult =pd.DataFrame({'day0' : [0]*self.user}) # 강화 성공 여부 매트릭스 
        self.upgradestate=pd.DataFrame({'day0':[0]*self.user}) # 강화 성공 상태 매트릭스
        self.failstate=pd.DataFrame({'day0':[0]*self.user}) # 강화 실패 누적 기록 매트릭스
        
        
    def addday(self): # 날짜를 다음날로 바꾸는 매소드 
    
        self.d = self.d+1
        
    def attempt(self): # 강화 성공 여부 저장 매소드  
        
        self.attemptresult['day%s' % self.d] = np.random.binomial(n=1, p= self.prob, size= self.user)
        
        row5= self.upgradestate.index[self.upgradestate['day%s'% (self.d-1)] == 5] # 강화 성공한 행 번호 
        
        
        self.attemptresult['day%s' % self.d][row5] = 1  # 강화 성공한 사람들은 다시 강화시도를 안한걸로 만들 수 있는 장치
        
        
    def update(self): # 강화시도결과를 upgrade state 와 failstate 에 저장하는 함수 
        
        self.upgradestate['day%s' % self.d]=self.upgradestate['day%s' % (self.d-1)]+self.attemptresult['day%s' % self.d] #강화 성공 업데이트 
        self.upgradestate = np.minimum(self.upgradestate,5) # 강화 5번 성공 이후 진행 X 
        
        
        self.failstate['day%s' % self.d] = self.failstate['day%s' % (self.d-1)]  
            
        rowX= self.upgradestate.index[self.upgradestate['day%s'%self.d] != 0 ] # 강화상태가 0이 아닌 행 번호 
        rowO= self.upgradestate.index[self.upgradestate['day%s'%self.d] == 0 ] # 강화상태가 0인 행 번호
            
        self.failstate['day%s' % self.d][rowX]=self.failstate['day%s' % (self.d-1)][rowX]+(self.attemptresult['day%s' % self.d][rowX]==0)*1 
        self.failstate['day%s' % self.d][rowO]=self.failstate['day%s' % (self.d-1)][rowO]
        
        
    def reset(self): # 누적실패가 2 일경우 강화상태를 초기화하는 함수 
        
        self.upgradestate['day%s' % self.d][self.failstate.index[self.failstate['day%s'%self.d] == 2]] = 0 
        self.failstate['day%s' % self.d][self.failstate.index[self.failstate['day%s'%self.d] == 2]] = 0
        
        #self.failstate.loc[self.failstate['day%s'%self.d] == 2, 'day%s'%self.d] == 0 
        
    def arrangeprob(self): # +5를 만든 유저가 50명이상일 경우 확률을 조정하는 함수 
        
        if sum(self.upgradestate['day%s' % self.d] == 5) >= 50:
            self.prob = 0.9
            
```

`-` 1번

```python
a=JPsord(0.1) # 확률 10% 
for i in range(62):
    a.addday()
    a.attempt()
    a.update()
```

```python
sum(a.upgradestate.day62==5) # 답
```

`-` 2번

```python
# prob 2-1, 2-2 
a=JPsord(0.7) # 확률 70%
for i in range(62):
    a.addday()
    a.attempt()
    a.update()
    a.reset()
```

```python
sum(a.upgradestate.day62==5) # 2-1 
```

```python
sum(a.upgradestate.day31==5) # 2-2
```

```python
# prob 2-3,
a=JPsord(0.7) ## 여기서 __init__() 실행
for i in range(62):
    a.addday()
    a.attempt()
    a.update()
    a.reset()
    a.arrangeprob()
```

```python
sum(a.upgradestate.day62==5) 
```
